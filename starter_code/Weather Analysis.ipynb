## Data Analysis

1. Cloud Coverage.   An odd trend with cloud coverage was the data aligning primarily on certain percentages.  This should not be considered a trend for analysis purposes.  Cloud coverage are determined on a 1/8 scale called an Otka Unit.  the sky is divided into 8 segments and if clouds are present in that grid that grid coordinate is considered " 100% full" regardless of what percentage for that grid is covered in cloud.  

2. At least with this data the equator does not appear to be the hottest in relation to other latitudes.  The current data is limited to a single day.  On that day 7/11/19 the hottest latitude was roughly 20 degrees north.   20 Degrees North correlates to the position in earth’s orbit where the tilt of the earth and the sun’s rays strike at a 90-degree angle.   Throughout the years this varies position varies from 23.5 degrees to -23.5 degrees also called the tropic of Capricorn and the tropic of Cancer respectively  

3. City Humidity is evenly distributed amongst the different Latitudes.  However, the humidity level median is around 60%. 

4. Windspeed.  There is a pattern that indicated wind speed is relatively higher at the equator.  this can be explained by the earths shape.  The earth is widest at the equator and because it is a round shape the equator travels faster than more southern and northern latitudes.   This extra energy and frictions with the air will cause higher wind speeds. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

wh_data = pd.read_csv("weather_data.csv")

wh_data.head()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
# Build a scatter plot for each data type
plt.scatter(wh_data["Lat"], wh_data["Max Temp"], marker="o", s=10)

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature")
plt.ylabel("Max. Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Max_Temp_vs_Latitude.png")

# Show plot
plt.show()
```

#### Latitude vs. Humidity Plot

```python
# Build a scatter plot for each data type

#my data had an anomolly of 290% humidity this code removes anomolous data points
wh_data_cl = wh_data.loc[wh_data['Humidity'] <= 101]

plt.scatter(wh_data_cl["Lat"], wh_data_cl["Humidity"], marker="o", s=10)

# Incorporate the other graph properties
plt.title("City Latitude vs. Humidity")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Humidity_vs_Latitude.png")

# Show plot
plt.show()

```

#### Latitude vs. Cloudiness Plot

```python
# Build a scatter plot for each data type
plt.scatter(wh_data["Lat"], wh_data["Cloudiness"], marker="o", s=10)

# Incorporate the other graph properties
plt.title("City Latitude vs. Cloudiness")
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Cloudiness_vs_Latitude.png")

# Show plot
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
# Build a scatter plot for each data type
plt.scatter(wh_data["Lat"], wh_data["Wind Speed"], marker="s", s=10)

# Incorporate the other graph properties
plt.title("City Latitude vs. Wind Speed")
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Wind_Speed_vs_Latitude.png")

# Show plot
plt.show()
```
